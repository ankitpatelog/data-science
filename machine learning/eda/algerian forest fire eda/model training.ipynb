```python
# function for finding the corr with the threshold and put it inside the set and remvoe it which not help to improve the model jsust makiing it overfit
def correlation(dataset,threshold):
    col_corr = set()
    corr_matrix = dataset.corr()
    for i in range(len(corr_matrix.columns)):
        # check for the prev columns 
        for j in range(i):
            if abs(corr_matrix[i,j]>threshold):
                col_corr.add(corr_matrix.columns[i])
        return col_corr
```
